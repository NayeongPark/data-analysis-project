## fillna() - 결측치 채우기

```python
df['age'] = df['age'].fillna(700)
df['age'].tail()
```

```python
df['embarked'].mode()
```

```python
df['embarked'].mode()[0]
```

```python
df['embarked'] = df['embarked'].fillna(df['embarked'].mode()[0])
df['embarked'].isnull().sum()
```

## dropna() - NaN 값이 있는 데이터 제거하기
- how = 'any'(기본값) // 하나라도 NaN이면 삭제
- how = 'all' // 모두 NaN이면 삭제

```python
df1 = df.copy()
df1.tail()
```

```python
df1.dropna()
```

```python
df1.dropna(how = 'all')
```

## Sort(정렬)

```python
df.sort_index(ascending = False).head() 
#index 기준으로 정렬
#ascending = False 로 내림차순으로 index 기준으로 출력
#ascending = True 오름차순 <-- 이게 기본값
```

```python
df.sort_values('age', ascending = False).head()
# 값을 기준으로 행 정렬
```

```python
df.sort_values(['fare', 'age'], ascending = False).head()
```

```python
df.sort_values(['fare', 'age'], ascending = [False, True]).head()
#fare 기준 내림차순/ fare 동률의 경우 age 기준 오름차순
```

## loc - indexing / slicing

- slicing은 [시작(포함): 끝(포함)]  

```python
df.loc[5, 'parch'] #indexing
```

```python
df.loc[2:5, ['sex','parch','fare']] #fancy indexing 가능
```

```python
df.loc[2:10:2, 'pclass':'embarked'] #열도 범위추출 가능.
```

- 조건별로 소괄호 꼭!!

```python
# boolean index 만들어 조건에 맞는 데이터만 추출 가능
cond = (df['pclass'] == 1)
cond
```

```python
df.loc[cond]
```

```python
# 다중 조건도 가능
cond1 = (df['sex'] == 'female')
cond2 = (df['pclass'] == 1)

df.loc[cond1 & cond2] #교집합
```

```python
df.loc[cond1 | cond2] #합집합
```

- 조건 필터 후 원하는 값 대입 가능

```python
df.loc[cond, 'age'] = -1
df.loc[cond]
```

## isin - 특정 값의 포함 여부 확인

```python
sample = pd.DataFrame({'name': ['kim', 'lee', 'park', 'choi'], 
                        'age': [24, 27, 34, 19]
                      })
sample
```

```python
sample['name'].isin(['kim'])
```

```python
cond2 = sample['name'].isin(['park','lee'])
sample.loc[cond2]
```

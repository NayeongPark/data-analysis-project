## json(JavaScript Object Notation) - 데이터 전송시 많이 사용됨

```python
import pprint
import json
import requests
```

```python
url = "https://api.exchangerate-api.com/v4/latest/USD" 
#json 형식의 파일을 로드하기 위해 가져온 url
```

```python
ret = requests.get(url)
#api로 실시간 데이터 요청
```

```python
json_data = json.loads(ret.text)
#json 형식의 파일로 로
```

```python
pprint.pprint(json_data)
#pprint = pretty print 보기 좋게 예쁘게 출력
```

## json 불러오기

```python
dff = pd.read_json(url)
dff
```

## json 형식으로 저장하기
